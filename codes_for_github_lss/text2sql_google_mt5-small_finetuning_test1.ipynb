```python
def extract_indexes(df): # DB Schema 정보를 불러온 DataFrame에서 사용할 정보만 뽑아옴
    """
    DataFrame에서 스키마 정보를 추출합니다.

    Args:
        df: 스키마 정보를 담고 있는 Pandas DataFrame.

    Returns:
        schema_info: 추출된 스키마 정보를 담은 리스트.
    """
    schema_info = []
    for index, row in df.iterrows():
        schema_info.append({
            "db_id": row['db_id'], #DB id
            "table_names_original": row['table_names_original'], #테이블 영문 이름
            "column_names_original": row['column_names_original'], #컬럼 영문 이름
            "column_types": row['column_types'] #컬럼 데이터 타입
        })
    return schema_info
```

```python
# DataFrame에서 스키마 정보 추출
schema_info = extract_indexes(df)

# 추출된 정보 출력
for schema in schema_info:
    print(schema)
```

```python
# 위에서 추출한 schema_info에서 db_id를 key, table_names_original과 column_names_original, column_types를 value로 해서 dictionary를 만듬

schema_dict = {}

for schema in schema_info:
  db_id = schema['db_id']
  schema_dict[db_id] = {
      'table_names_original': schema['table_names_original'][0],
      'column_names_original': [col[1] for col in schema['column_names_original']],
      'column_types': schema['column_types']
  }
```

```python
# 위에서 만든 dictionary 내용 확인

for key, value in schema_dict.items():
  print(f"Key: {key}")
  print(f"Value: {value}")
  print("-" * 20)
```

```python
# json 파일 하나씩 불러와서 하나의 DataFrame에 합침

def merge_json_files_to_dataframe(root_dir):
    """
    Merges all JSON files in subfolders of the root directory into a single DataFrame.

    Args:
        root_dir: The path to the root directory.

    Returns:
        A Pandas DataFrame containing the merged data.
    """
    df = pd.DataFrame()  # Create DataFrame

    for dirpath, dirnames, filenames in os.walk(root_dir):
        for filename in filenames:
            if filename.endswith('.json'):
                filepath = os.path.join(dirpath, filename)
                print(filepath)
                with open(filepath, 'r', encoding='utf-8') as f:
                    try:
                        data = json.load(f)
                        df_temp = pd.DataFrame(data["data"])
                        df = pd.concat([df, df_temp])
                    except json.JSONDecodeError as e:
                        print(f"Error decoding JSON in file: {filepath}")
                        print(f"Error message: {e}")

    return df
```

```python
# Training Data Set의 라벨링 데이터가 있는 구글 드라이브 폴더 경로를 지정합니다.
folder_path = '/content/drive/MyDrive/Project1/Data/Training/SQLs'

# JSON 파일을 읽어와 병합합니다.
df2 = merge_json_files_to_dataframe(folder_path)

# 병합된 DataFrame을 출력합니다.
print(df2.info())
```

## 2. Pytorch 데이터 로더 세팅

```python
import torch
from torch.utils.data import Dataset, DataLoader
```

```python
# 모델 학습에 사용할 Dataset 형식 생성

class SQLDataset(Dataset):
    def __init__(self, df2, schema_dict):
        self.df2 = df2
        self.schema_dict = schema_dict

    def __len__(self):
        return len(self.df2)

    def __getitem__(self, idx):
        row = self.df2.iloc[idx]
        db_id = row['db_id'] #DB schema 정보를 dictionary로 부터 불러오기 위해 사용할 key
        utterance = row['utterance'] #자연어 질문
        query = row['query'] #결과 쿼리

        if db_id in self.schema_dict:
            table_names = self.schema_dict[db_id]['table_names_original'] # 테이블 영문 이름
            column_names = self.schema_dict[db_id]['column_names_original'] # 컬럼 영문 이름
            column_types = self.schema_dict[db_id]['column_types'] # 컬럼 데이터 타입

            # 여기서 토큰화 및 인코딩 작업을 수행합니다.
            # 예를 들어, utterance, table_names, column_names를 토큰화하고 인덱스로 변환합니다.
            # input_tokens = tokenize_and_encode(utterance, table_names, column_names)

            # 임시로 문자열을 그대로 사용
            input_tokens = f"[CLS]{utterance}[SEP]{table_names}{'[SEP]'.join(column_names)}" #input 형식

            # query를 토큰화하고 인덱스로 변환합니다.
            # target_tokens = tokenize_and_encode(query)

            # 임시로 문자열을 그대로 사용
            target_tokens = query #결과 쿼리

            return input_tokens, target_tokens
        else:
            return None, None
```

```python
# 데이터 로더 생성
dataset = SQLDataset(df2, schema_dict)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
```

```python
# 데이터 로더 사용 예시
for input_tokens, target_tokens in dataloader:
    # 모델에 input_tokens를 입력하고 target_tokens를 정답으로 사용하여 학습합니다.
    print(input_tokens[0])
    print(target_tokens[0])
    break # 예시로 첫 번째 배치만 출력
```

## 3. Load Pretrained Model

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
tokenizer = AutoTokenizer.from_pretrained("google/mt5-small")
model = AutoModelForSeq2SeqLM.from_pretrained("google/mt5-small")
```

```python
# 위에서 불러온 모델을 파인튜닝해서 text2sql task를 수행하는 모델이 되도록 파인튜닝.
# 학습 data는 위에서 만든 data loader
# 학습할때마다 loss와 정확도를 출력

from transformers import AdamW
from tqdm import tqdm

# Hyperparameters
epochs = 3
learning_rate = 5e-5
# 배치 크기 감소
batch_size = 8  # 또는 더 작게, 4, 2 등을 시도해 보세요.
# 그래디언트 누적 단계
gradient_accumulation_steps = 4  # 이렇게 하면 배치 크기가 32(8 * 4)가 됩니다.

# Optimizer
optimizer = AdamW(model.parameters(), lr=learning_rate)

# Move model to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# DataLoader 업데이트 - 더 작은 배치 크기 사용
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# Training loop
for epoch in range(epochs):
    total_loss = 0
    correct_predictions = 0
    total_predictions = 0

    progress_bar = tqdm(dataloader, desc=f"Epoch {epoch + 1}/{epochs}")
    for input_tokens, target_tokens in progress_bar:
        # Tokenize and encode the input and target sequences
        inputs = tokenizer(input_tokens, return_tensors="pt", padding=True, truncation=True).to(device)
        targets = tokenizer(target_tokens, return_tensors="pt", padding=True, truncation=True).to(device)

        # Forward pass
        outputs = model(**inputs, labels=targets.input_ids)
        loss = outputs.loss

        # Backward pass and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Calculate accuracy
        predicted_ids = torch.argmax(outputs.logits, dim=-1)
        correct_predictions += (predicted_ids == targets.input_ids).sum().item()
        total_predictions += targets.input_ids.numel()

        total_loss += loss.item()

        progress_bar.set_postfix({"loss": loss.item(), "accuracy": correct_predictions / total_predictions})

    average_loss = total_loss / len(dataloader)
    accuracy = correct_predictions / total_predictions
    print(f"Epoch {epoch + 1}/{epochs}, Loss: {average_loss}, Accuracy: {accuracy}")

# Save the fine-tuned model
# model.save_pretrained("fine_tuned_text2sql_model")
```

```python
# 모델 저장 전에 가중치를 연속적으로 복사
for name, param in model.named_parameters():
    param.data = param.data.contiguous()

# 모델 저장
save_path = '/content/drive/MyDrive/Project1/fine_tuned_mt5-small_model'  # 모델 저장할 경로 지정
os.makedirs(save_path, exist_ok=True)
model.save_pretrained(save_path)
tokenizer.save_pretrained(save_path)

print(f"Model saved to: {save_path}")
```
